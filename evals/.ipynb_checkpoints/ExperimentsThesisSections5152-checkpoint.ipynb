### Functions running the set of experiments using SafeOpt and LB_SGD

```python
def run_exp_SafeOpt(f, h, 
                    d, m,
                    experiments_num = 5, 
                    n_iters = 100,  
                    x00 = np.zeros(2), 
                    x_opt =  np.ones(2) / 2**0.5,  
                    sigma = 0.001,                # noise in measurements
                    bnd_l = -1., bnd_u = 1.,      # bounds of sampling for SafeOpt
                    problem_name = '',
                    gp_var = 0.01, L = 0.25,  gp_num_samples = 50 ):

    errors = []
    cons = []
    runtimes = []
    for i in range(experiments_num):
        time_0 = time()
        x_traj, errors_so, cons_so, gp = run_SafeOpt(n_iters, 
                                                     f, h, 
                                                     x00, x_opt, 
                                                     d, m, 
                                                     sigma, bnd_l, bnd_u, gp_var, L = L,  
                                                     gp_num_samples = gp_num_samples)
        runtimes.append(time() - time_0)
        print("SafeOptRun finished")
        errors.append(errors_so)
        cons.append(cons_so)
    errors = np.array(errors)
    cons = np.array(cons)
    runtimes = np.array(runtimes)
    
    with open('../runs/SafeOpt_' + problem_name + '_d' + str(d)  + '.npy', 'wb') as file:
        np.save(file, errors)
        np.save(file, cons)
        np.save(file, runtimes)
    
    return (errors, cons, gp)  

def run_exp_LB_SGD(f, h, d, m,
                   experiments_num = 5, 
                   n_iters = 100, 
                   n = 1, 
                   M0 = 0.5 / 2., 
                   Ms = 0. * np.ones(4), 
                   x00 = np.zeros(2), 
                   x_opt =  np.ones(2) / 2**0.5, 
                   sigma = 0.001, nu = 0.01, 
                   eta0 = 0.05, 
                   T = 3, 
                   factor = 0.85, 
                   init_std = 0.1,
                   delta = 0.01,
                   problem_name = ''):

    my_oracle = LB.Oracle(
        f = f,
        h = h, 
        sigma = sigma,
        hat_sigma = 0.01,
        delta = delta,
        m = m,
        d = d,
        nu = nu,
        zeroth_order = True,
        n = n,
        Ms = Ms)

    opt = LB.SafeLogBarrierOptimizer(
        x00 = x00,
        x0 = x00,
        M0 = M0,
        Ms = Ms,
        sigma = my_oracle.sigma,
        hat_sigma = my_oracle.hat_sigma,
        init_std = init_std,
        eta0 = eta0,
        oracle = my_oracle,
        f = f,
        h = h,
        d = d,
        m = m,
        reg = 0.0001,
        x_opt = x_opt,
        factor = factor,
        T = T,
        K = int(n_iters / T / 2. / n),
        experiments_num = experiments_num,
        mu = 0.,
        convex = True,
        random_init = True,
        no_break = True)

    opt.run_average_experiment()
    
    for i in range(experiments_num):
        opt.errors_total[i] = np.repeat(opt.errors_total[i], 2 * n)
        opt.constraints_total[i] = np.repeat(opt.constraints_total[i], 2 * n )
    
    errors = opt.errors_total
    constraints = opt.constraints_total
    runtimes = opt.runtimes
    runtimes = np.array(runtimes)
    
    with open('../runs/LB_SGD_' + problem_name + '_d' + str(d)  + '.npy', 'wb') as file:
        np.save(file, errors)
        np.save(file, constraints)
        np.save(file, runtimes)

    return opt
```

## Section 5.1 Cutting machine parameters tuning

### Defining the problem

```python
d = 2
m = 2 * d + 1
experiments_num = 10
x_opt = np.array([0.1634, 0.131])
M0 = 100.
Ms = np.array([25.,0.,0.,0.,0.])
T = 30
sigma = 0.001
L = 2.

aT = 127.5365
bT = np.array([-846.29, -144.2])
cT = np.array([1703, 365.6])

def Tf(x):
    return aT + bT[0] * x[0] + bT[1] * x[1] + cT[0] * x[0]**2 + cT[1] * x[1] * x[0]

aR = 0.9094 
bR = np.array([-10.035, 7.0877])
cR = np.array([34., -18.969])
Lf = 300. #mm
Df = 41. #mm

def R(x):
    return aR + bR[0] * x[0] + bR[1] * x[1] + cR[0] * x[0]**2 + cR[1] * x[1] * x[0]

C_I = 50.
C_M = 40.

def t_c(x):
    return Lf * Df * np.pi / (x[0] * x[1] * 1e6 / 60.)

def f(x):
    Tff = Tf(x) / 60.
    return t_c(x) * (C_M / 3600. + C_I / Tff / 3600.)

def h(x):
    m = 5
    constr = np.zeros(m)
    factor = 10.
    constr[0] = factor * (x[0] - 0.2)
    constr[1] = factor * (-x[0] + 0.1)
    constr[2] = factor * (x[1] - 0.16)
    constr[3] = factor * (-x[1] + 0.08)
    constr[4] = - 0.7 + R(x)
    return constr
fc = f
```

```python
from lib.plot_thesis import plot_dyn_trajectory
from lib.dyns import dyn_linear, fgamma, dyn_heli_linearized, dyn_unicycle
from lib.optimization_functions_dyn_thesis import df_z, g_obstacle, f, g_round_wall, g_straight_walls
import scipy

from lib.SFW import get_measurements, make_and_add_measurement, estimate_constraints, solve_DFS, true_solution, violation, SCFW, SCFW_reduced, Exp
```

### Initial stationary trajectory

```python
# Linear dynamics 

N = 10
dt = 0.1

x_0 = np.array([-2.,0.])
x_C = np.array([0.,0.])
x_B = np.array([2.5,0.])
dyn = dyn_linear
A = np.array([
            [1., 0.5],
            [0., 1.]
                ])

B = np.array([0., 1.])
p = 1
args = (x_0, x_B, x_C, A, B, N, dt)

loop = 'open'
u = np.zeros(N)
R = 3
Aw = np.array([[1,1],[-1,-1]])
bw = np.array([3,3])
# Aw2 = np.array([[1,-1],[1,-1]])
# bw2 = np.array([3,3])
# print(f(u, args, loop, dyn_linear))
x = dyn_linear(u, args, loop)
plot_dyn_trajectory(x, args, 
                    xlim = (-4, 4), 
                    ylim = (-4, 4), 
                    constraint = 'convex', 
                    R = R, 
                    b1 = bw[0], 
                    b2 = bw[1])
```

### Optimizing the controller using scipy

```python
def f_u(x):
    U = np.reshape(x, (N,p))
    return f(U, args, loop, dyn)

def h_u(x):
    U = np.reshape(x, (N,p))
    g_round =  g_round_wall(U, R, args, loop, dyn)
    g_straight =  g_straight_walls(U, Aw, bw, args, loop, dyn).reshape((N-1) * np.size(bw))
    return np.hstack((g_round , 
                     g_straight))

#Optimal
fun = lambda x: f_u(x)
cons = ({'type': 'ineq', 'fun': lambda x: -h_u(x)})
# nlc = scipy.optimize.NonlinearConstraint(cons, -np.inf, 0)
# ,{'type': 'ineq', 'fun': lambda x: -x.dot(A3.dot(x)) - b3.dot(x) - c3})

u_opt = scipy.optimize.minimize(f_u, np.zeros(N),
                                method = 'SLSQP',  
                                constraints=cons, 
                                tol = 0.0001)
u_opt = u_opt.x
x_opt = dyn(u_opt, args, loop = 'open')
plot_dyn_trajectory(x_opt, args, 
                    xlim = (-4, 4), ylim = (-4, 4), 
                    constraint = 'convex', 
                    R = R, b1 = bw[0], b2 = bw[1], 
                    problem_name = 'LQR_linear_init')

#BoundaryPoint
x_B = np.array([2.5,0.])
args = (x_0, x_B, x_C, A, B, N, dt)
eta0 = 0.25
experiments_num = 2
x_opt = u_opt
u00 = np.reshape(u, N*p)
dyn = dyn_linear
sigma = 0.001
de = 0.01

n_iters = 1500
n = 6
d = N * p
m = np.size(h_u(u00))
experiments_num = 10
M0 = 1.
Ms = 1. * np.ones(m)
sigma = 0.001
problem_name = 'LQR_linear'
```
